## 矩阵求导公式的数学推导

### 公式1

$$
\frac{\partial\left(\boldsymbol{x}^{T} \boldsymbol{a}\right)}{\partial \boldsymbol{x}}=\frac{\partial\left(\boldsymbol{a}^{T} \boldsymbol{x}\right)}{\partial \boldsymbol{x}}=\boldsymbol{a} \tag{1}
$$

其中，$\boldsymbol{a}$ 为常数向量，$ \boldsymbol{a}=\left(a_{1}, a_{2}, \cdots, a_{n}\right)^{T} $ 。


证明：

因为

$$
\left [ a1, a2,\dots ,an \right ] \cdot \begin{bmatrix}
x_{1} \\
x_{2}\\
\vdots  \\
x_{n}
\end{bmatrix} = \sum_{i=1}^{n}a_{i}x_{i} \tag{2}
$$



所以

$$
\begin{aligned}
\frac{\partial\left(\boldsymbol{x}^{T} \boldsymbol{a}\right)}{\partial \boldsymbol{x}} &=\frac{\partial\left(\boldsymbol{a}^{T} \boldsymbol{x}\right)}{\partial \boldsymbol{x}} \\
&=\frac{\partial\left(a_{1} x_{1}+a_{2} x_{2}+\cdots+a_{n} x_{n}\right)}{\partial \boldsymbol{x}} \\
&=\left[\begin{array}{l}
\frac{\partial\left(a_{1} x_{1}+a_{2} x_{2}+\cdots+a_{n} x_{n}\right)}{\partial x_{1}} \\
\frac{\partial\left(a_{1} x_{1}+a_{2} x_{2}+\cdots+a_{n} x_{n}\right)}{\partial x_{2}} \\
\vdots \\
\frac{\partial\left(a_{1} x_{1}+a_{2} x_{2}+\cdots+a_{n} x_{n}\right)}{\partial x_{n}}
\end{array}\right] \\
&=\left[\begin{array}{l}
a_{1} \\
a_{2} \\
\vdots \\
a_{n}
\end{array}\right] \\
&=\boldsymbol{a}
\end{aligned} \tag{3}
$$

证毕。

### 公式2

$$
\frac{\partial\left(\boldsymbol{x}^{T} \boldsymbol{x}\right)}{\partial \boldsymbol{x}}=2 \boldsymbol{x} \tag{4}
$$ 

因为:

$$
\left [ x_{1}, x_{2},\dots ,x_{n}\right ] \cdot \begin{bmatrix}
x_{1} \\
x_{2}\\
\vdots  \\
x_{n}
\end{bmatrix} = \sum_{i=1}^{n}x_{i}^2 \tag{5}
$$


证明：

$$
\begin{aligned}
\frac{\partial\left(\boldsymbol{x}^{T} \boldsymbol{x}\right)}{\partial \boldsymbol{x}} &=\frac{\partial\left(x_{1}^{2}+x_{2}^{2}+\cdots+x_{n}^{2}\right)}{\partial \boldsymbol{x}} \\
&=\left[\begin{array}{c}
\frac{\partial\left(x_{1}^{2}+x_{2}^{2}+\cdots+x_{n}^{2}\right)}{\partial x_{1}} \\
\frac{\partial\left(x_{1}^{2}+x_{2}^{2}+\cdots+x_{n}^{2}\right)}{\partial x_{2}} \\
\vdots \\
\frac{\partial\left(x_{1}^{2}+x_{2}^{2}+\cdots+x_{n}^{2}\right)}{\partial x_{n}}
\end{array}\right] \\
&=\left[\begin{array}{c}
2 x_{1} \\
2 x_{2} \\
\vdots \\
2 x_{n}
\end{array}\right] \\
&=2\left[\begin{array}{c}
x_{1} \\
x_{2} \\
\vdots \\
x_{n}
\end{array}\right] \\
&=2 \boldsymbol{x}
\end{aligned} \tag{6}
$$

### 公式3

$$
\frac{\partial\left(\boldsymbol{x}^{T} \boldsymbol{A} \boldsymbol{x}\right)}{\partial \boldsymbol{x}}=\boldsymbol{A} \boldsymbol{x}+\boldsymbol{A}^{T} \boldsymbol{x} \tag{6}
$$ 

其中，$\boldsymbol{A}_{n\times n}$ 是常数矩阵，$\boldsymbol{A}_{n \times n}=\left(a_{i j}\right)_{i=1, j=1}^{n, n}$ 。

$$
\left [ x_{1}, x_{2}, \dots  ,x_{n}\right ] \begin{bmatrix}
 a_{11} &  a_{12}  &  \dots & a_{1n}  \\
   a_{21} &  a_{22}  &  \dots & a_{2n}  \\
  \vdots & \vdots  &  \dots & \vdots \\
   a_{n1} &  a_{n2}  &  \dots & a_{nn} 
\end{bmatrix}\begin{bmatrix}
 x_{1}\\ x_{2}
\\ \vdots 
 \\ x_{n} 
\end{bmatrix} = \sum_{i=1}^{n} \sum_{j=1}^{n} a_{ij}x_{i}x_{j} \tag{7}
$$

证明：

$$
\frac{\partial(x^TAx)}{\partial x} = \frac{\partial (a_{11}x_{1}x_{1}+a_{12}x_{1}x_{2}+\cdots +a_{1n}x_{1}x_{n} \\ +a_{21}x_{2}x_{1}+a_{22}x_{2}x_{2}+\cdots +a_{2n}x_{2}x_{n} \\ + \cdots \\ +a_{n1}x_{n}x_{1}+a_{n2}x_{n}x_{2}+\cdots +a_{nn}x_{n}x_{n}  ) }{\partial x}  
$$

$$

### 公式4

$$
\frac{\partial\left(\boldsymbol{a}^{T} \boldsymbol{x} \boldsymbol{x}^{T} \boldsymbol{b}\right)}{\partial \boldsymbol{x}}=\boldsymbol{a} \boldsymbol{b}^{T} \boldsymbol{x}+\boldsymbol{b a}^{T} \boldsymbol{x} \tag{9}
$$ 

其中，$a,b$ 为常数向量，$a = (a_{1}, a_{2}, \cdots, a_{n})^T, b=(b_{1},b_{2},\cdots, b_{n})^T$。

证明：

因为 $a^Tx=x^Ta, x^Tb=b^Tx$，所以：

$$
\frac{\partial (a^Txx^Tb)}{\partial  x}=\frac{\partial (x^Tab^Tx)}{\partial x}  \tag{10}
$$

又因为 $ab^T$ 是 $n \times n$ 常数矩阵，是（6）式得:

$$
\frac{\partial (a^Txx^Tb)}{\partial  x}=\frac{\partial (x^Tab^Tx)}{\partial x}  =ab^Tx+ba^Tx \tag{11}
$$

## 最小乘法举例

$$
Y=\begin{bmatrix}
 y_{1}\\
  y_{2}\\
  \vdots \\
 y_{n}
\end{bmatrix} \tag{12}  
$$

其中，$Y$ 是 $n$ 行 $1$ 列。

$$
X=\begin{bmatrix}
 x_{1}^T\\
  x_{2}^T\\
  \vdots \\
 x_{n}^T
\end{bmatrix} \tag{13}
$$

其中，$x_{1}^T=[x_{11}, x_{12}, \cdots, x_{1p} ]$

$$
b=\begin{bmatrix}
 b_{1}\\
  b_{2}\\
  \vdots \\
 b_{p}
\end{bmatrix} \tag{14} 
$$




$$
L(b)=\sum_{i=1}^{n}\left(y_{i}-\left(x_{i}^{\top} \cdot b\right)\right)^{2} \tag{15} 
$$

$$
L(b)=(Y-X b)^{\top} \cdot(Y-X b)\tag{16} 
$$

对$b$ 求偏导，$\frac{\partial L(b)}{\partial b} =0$

所以:

$$
L(b)=(Y-X b)^{\top} \cdot(Y-X b) \\ 
    = (Y^T -b^Tx^T) \cdot(Y-xb) \\
    = Y^T Y - Y^Txb - b^Tx^TY + b^Tx^T xb \\
    = Y^TY-2Y^Txb + b^Tx^Txb  \tag{17} 
$$


因为 $Y^Txb$ 是一个标量（$1\times n \cdot n \times p \cdot p \times 1 = 1 \times 1$），所以$ Y^Txb = b^Tx^TY $ 

下面对公式17的三项分别求导：

$$
\frac{\partial (Y^TY)}{\partial b} =\begin{bmatrix}
 0\\
 0\\
 \vdots \\
0
\end{bmatrix} \tag{18} 
$$ 

$$
\frac{2\partial (Y^Txb)}{\partial b} = 2(Y^Tx)^T=2x^TY \tag{19} 
$$

$$
\frac{\partial (b^Tx^Txb)}{\partial b} = (x^Tx+x^Tx)\cdot b=2x^Tx \cdot b \tag{20} 
$$

继续对公式17求导：
$$
\frac{\partial L(b)}{\partial b} = -2x^TY+2x^Txb=0 \tag{21}
$$

所以：
$$
\hat b=(x^Tx)^{-1} \cdot x^T \cdot Y
$$
